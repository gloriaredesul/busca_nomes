```python
# üìå Passo 1: Importar bibliotecas necess√°rias
import pandas as pd
from datetime import datetime
from google.colab import files
import io

# üìå Passo 2: Upload da planilha
print("üìÅ Fa√ßa o upload da planilha (.xlsx ou .csv):")
uploaded = files.upload()

# üìå Passo 3: Leitura da planilha
filename = list(uploaded.keys())[0]

if filename.endswith('.xlsx'):
    df = pd.read_excel(io.BytesIO(uploaded[filename]))
elif filename.endswith('.csv'):
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
else:
    raise ValueError("Formato de arquivo n√£o suportado. Envie .xlsx ou .csv")

# üìå Passo 4: Convers√£o das datas (mantendo colunas com inicial mai√∫scula)
df['Vencimento'] = pd.to_datetime(df['Vencimento'], errors='coerce')
df['Pagamento'] = pd.to_datetime(df['Pagamento'], errors='coerce')

# üìå Passo 5: Verifica√ß√£o de pagamentos em dia (mesmo m√™s e ano)
df['pagamento_em_dia'] = (
    (df['Vencimento'].dt.month == df['Pagamento'].dt.month) &
    (df['Vencimento'].dt.year == df['Pagamento'].dt.year)
)

# üìå Passo 6: Filtrar apenas os pagamentos em dia
pagamentos_em_dia = df[df['pagamento_em_dia']]

# üìå Passo 7: Exibir resultado
print("‚úÖ Pagamentos realizados no mesmo m√™s e ano do vencimento:")
display(pagamentos_em_dia)

```
